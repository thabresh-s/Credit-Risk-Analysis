```python
# Step 2: Import the necessary libraries
import pandas as pd
import quantstats as qs
import yfinance as yf
```

```python
# Step 3: Load the TCS stock data into a Pandas DataFrame using yfinance
symbol = "TCS.NS"  # TCS stock symbol on the National Stock Exchange of India
data = yf.download(symbol, start="2015-01-01", end="2021-12-31")
```

```python
# Step 4: Calculate the performance metrics using the QuantStats library
qs.extend_pandas()
returns = data['Close'].pct_change()
qs.reports.html(returns, output='tcs_report.html', title='TCS Stock Report')
```
